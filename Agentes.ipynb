```python
# ═══ Meu Primeiro Agente — O Aspirador de Pó ═══
import random

class Ambiente:
    """Ambiente com N salas que podem estar limpas (0) ou sujas (1)."""
    def __init__(self, n_salas=2):
        self.salas = [random.choice([0, 1]) for _ in range(n_salas)]
        self.pos = 0              # posição do agente
        self.score = 0
        self.historico = []

    def perceber(self):
        """O agente percebe: (posição, está_sujo?)"""
        return (self.pos, self.salas[self.pos])

    def executar(self, acao):
        """Executa a ação e atualiza o score."""
        self.score -= 1  # custo de energia
        if acao == "aspirar":
            self.salas[self.pos] = 0
        elif acao == "direita" and self.pos < len(self.salas) - 1:
            self.pos += 1
        elif acao == "esquerda" and self.pos > 0:
            self.pos -= 1
        # Bônus: +10 por cada sala limpa
        self.score += self.salas.count(0) * 10
        self.historico.append({'pos': self.pos, 'salas': self.salas[:], 'score': self.score})

# ─── Agente Reativo Simples ───
def agente_reativo(percepcao):
    """Se sujo, aspira. Se não, move para o outro lado."""
    pos, sujo = percepcao
    if sujo:
        return "aspirar"
    elif pos == 0:
        return "direita"
    else:
        return "esquerda"

# ─── Rodar ───
env = Ambiente(n_salas=5)
print(f"🏠 Estado inicial: salas = {env.salas}, agente na sala {env.pos}\n")

for i in range(6):
    p = env.perceber()
    a = agente_reativo(p)
    env.executar(a)
    status = ['🟩' if s == 0 else '🟫' for s in env.salas]
    cursor = [' '] * len(env.salas)
    cursor[env.pos] = '🤖'
    print(f"Passo {i+1}: percepção={p} → ação={a:<10s} salas={''.join(status)} agente={''.join(cursor)} score={env.score}")

print(f"\n📊 Score final: {env.score}")
```

```python
# ═══ Ambiente Dinâmico — salas sujam sozinhas ═══

class AmbienteDinamico(Ambiente):
    """Ambiente onde salas podem ficar sujas aleatoriamente."""

    def __init__(self, n_salas=5, prob_sujeira=0.3):
        super().__init__(n_salas)
        self.prob_sujeira = prob_sujeira

    def executar(self, acao):
        super().executar(acao)
        # Depois da ação, salas podem ficar sujas de novo!
        for i in range(len(self.salas)):
            if self.salas[i] == 0 and random.random() < self.prob_sujeira:
                self.salas[i] = 1

# Rodar por mais passos para ver o efeito
env2 = AmbienteDinamico(n_salas=5, prob_sujeira=0.1)
print(f"🏠 Estado inicial: {env2.salas}\n")

scores = []
for i in range(30):
    p = env2.perceber()
    a = agente_reativo(p)
    env2.executar(a)
    scores.append(env2.score)
    if (i+1) % 5 == 0:
        status = ['🟩' if s==0 else '🟫' for s in env2.salas]
        print(f"Passo {i+1:2d}: {''.join(status)} pos={env2.pos} score={env2.score}")

# Gráfico do score
plt.figure(figsize=(10, 4))
plt.plot(scores, color='#00d4aa', linewidth=2)
plt.title('Score do Agente Reativo em Ambiente Dinâmico')
plt.xlabel('Passo'); plt.ylabel('Score acumulado')
plt.grid(alpha=0.3); plt.show()
```

```python
# ═══ Comparação: Aleatório vs Reativo vs Com Modelo ═══

# 1. Agente Aleatório (baseline ruim)
def agente_aleatorio(percepcao):
    """Escolhe ação aleatória. Usado como baseline."""
    return random.choice(["aspirar", "esquerda", "direita"])

# 2. Agente Reativo (já temos)
# agente_reativo já definido acima

# 3. Agente com Modelo — lembra quais salas já limpou!
class AgenteComModelo:
    """Mantém estado interno: mapa de salas conhecidas."""
    def __init__(self, n_salas):
        # modelo interno: None = desconhecido, 0 = limpo, 1 = sujo
        self.modelo = [None] * n_salas
        self.n_salas = n_salas

    def decidir(self, percepcao):
        pos, sujo = percepcao
        # Atualiza modelo interno
        self.modelo[pos] = sujo

        if sujo:
            return "aspirar"

        # Procura sala desconhecida ou suja no modelo
        # Primeiro à direita, depois à esquerda
        for i in range(pos + 1, self.n_salas):
            if self.modelo[i] is None or self.modelo[i] == 1:
                return "direita"
        for i in range(pos - 1, -1, -1):
            if self.modelo[i] is None or self.modelo[i] == 1:
                return "esquerda"

        # Tudo limpo e conhecido? Fica parado (ação sem efeito)
        return "direita" if pos == 0 else "esquerda"


# ─── Simulação comparativa ───
def simular(agente_fn, n_salas=5, n_passos=50, prob_sujeira=0.05, n_rodadas=100):
    """Roda várias simulações e retorna média dos scores."""
    todos_scores = []
    for _ in range(n_rodadas):
        env = AmbienteDinamico(n_salas, prob_sujeira)
        for _ in range(n_passos):
            p = env.perceber()
            a = agente_fn(p)
            env.executar(a)
        todos_scores.append(env.score)
    return np.mean(todos_scores), np.std(todos_scores)

# Wrapper para o agente com modelo
def criar_agente_modelo(n_salas):
    agente = AgenteComModelo(n_salas)
    return agente.decidir

n = 5
resultados = {}
for nome, fn in [("Aleatório", agente_aleatorio),
                    ("Reativo", agente_reativo),
                    ("Com Modelo", criar_agente_modelo(n))]:
    media, desvio = simular(fn, n_salas=n)
    resultados[nome] = (media, desvio)
    print(f"{nome:<12s}: score médio = {media:8.1f} ± {desvio:.1f}")

# ─── Gráfico de barras ───
nomes = list(resultados.keys())
medias = [resultados[n][0] for n in nomes]
desvios = [resultados[n][1] for n in nomes]
cores = ['#ef4444', '#3b82f6', '#00d4aa']

plt.figure(figsize=(8, 5))
bars = plt.bar(nomes, medias, yerr=desvios, color=cores, capsize=5, edgecolor='white', linewidth=0.5)
plt.title('Comparação de Agentes (100 simulações, 5 salas)', fontsize=14, fontweight='bold')
plt.ylabel('Score médio')
plt.grid(axis='y', alpha=0.3)
for bar, m in zip(bars, medias):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
             f'{m:.0f}', ha='center', fontweight='bold')
plt.tight_layout(); plt.show()
```
